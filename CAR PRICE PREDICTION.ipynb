# 4.Training the model

```python
x = sample.drop(columns=["price","CarName","fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem",])
y = sample["price"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 0)
```

```python
print('Training X Shape:', x_train.shape)
print('Training Y Shape:', y_train.shape)
print('Testing X Shape:', x_test.shape)
print('Testing Y Shape:', y_test.shape)
```

```python
model=RandomForestRegressor()
```

```python
rfe = RandomForestRegressor(n_estimators=40,max_depth=20)
rfe = rfe.fit(x, y)
```

# 5.Predictions

```python
predictions = rfe.predict(x_test)
```

```python
errors = abs(predictions - y_test)
```

```python
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

# 6.Model Evaluation

```python
mape = 100 * (errors / y_test)
```

```python
accuracy = 100 - np.mean(mape)
print('Accuracy using RandomForestRegressor:', round(accuracy, 2), '%.')
```
